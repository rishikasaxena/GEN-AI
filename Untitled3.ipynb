```python
from openai import OpenAI
client = OpenAI(api_key = "YOUR_API_KEY_HERE")
```

```python
messages = [ # Defines an AI timetable assistant restricted to generating structured study and placement schedules only.

    {"role": "system",
        "content": '''you are a timetable assistant for 4th year B.Tech students. You ONLY create structured daily or weekly timetables related to study, placements , workshops, and productivity. If the user asks anything else, reply: 'Sorry, I can only help with timetable-related questions.'''},
    {
        "role": "user",
        "content": "What is machine learning?"
    }
]
```

```python
response = client.chat.completions.create(
  model="gpt-3.5-turbo-16k",
    messages=messages,
  max_tokens=500
)
```

```python
print(response.choices[0].message.content)
```
