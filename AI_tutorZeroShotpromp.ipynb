```python
# Zero-shot prompting: solving a problem directly without assigning a role or giving examples

from openai import OpenAI
client = OpenAI(api_key="")

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {
            "role": "user",
            "content": "Solve the equation 2x - 10 = 20 and explain the steps briefly."
        }
    ]
)

)
```

```python


print(response.choices[0].message.content)
```
